```python
def uniq_cols(frame):
    t=[]
    for i in range(len(frame)):
        t.append(frame[i].columns)
    r=set()   
    for j in t:
        for k in j:
            r.add(k)

    l=[]
    cols=[]
    for g in r:
        l.append(g)

    for h in l:
        if type(h)==str:
            cols.append(h)
    return cols
        
```

```python
def temp_uniq(comp):
    column=uniq_cols(comp)
    
    t = re.compile("THER*")
    ther_list = set(filter(t.search,column))
    ther_list

    te = re.compile("TEMP*")
    temp_list = set(filter(te.search,column))
    temp_list
    return set.union(temp_list,ther_list)

temp_uniq(comp)
```

```python
def isNAN(val):
    return val!=val
```

```python
#columns with 'temp' or 'ther' in the name has name issue while reading the files
col_issue = set()
for df in range(len(comp)):
    intersection = list(ther_list & set(comp[df].columns))+list(temp_list & set(comp[df].columns))
    if intersection == 0:
        pass
    else:
        for i in range(len(intersection)):
            if isinstance(list(comp[df][intersection[i]])[0],str) == True:
            #if isNAN(list(comp[df][intersection[i]])[0]) == False:
                col_issue.add(df)

col_issue

temp_col_all = temp_uniq(comp)
#data frame with NaN columns after temprature column
nan_issue=set()
for df in range(len(comp)):
    for i in range(len(comp[df].columns)):
        if comp[df].columns[i] in temp_col_all:
            if i < len(comp[df].columns)-1:
                if isNAN(comp[df].columns[i+1]):
                    nan_issue.add(df)
nan_issue

cols_iss_all = set.union(col_issue, nan_issue)
cols_iss_all
```
